```python
import yfinance as yf
import matplotlib.pyplot as plt

# Get Tesla stock data
tesla = yf.Ticker("TSLA")
tesla_data = tesla.history(period="5y")  # For example, last 5 years

# Define the function to plot the stock graph
def make_graph(data, title):
    plt.figure(figsize=(10,6))
    plt.plot(data.index, data['Close'], label='Tesla Stock Price', color='blue')
    plt.title(title)
    plt.xlabel('Date')
    plt.ylabel('Stock Price (USD)')
    plt.grid(True)
    plt.legend()
    plt.show()

# Plot Tesla stock data
make_graph(tesla_data, "Tesla Stock Price Over the Last 5 Years")
```
